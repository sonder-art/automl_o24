### Preguntas Finales

```python
# Explica los conceptos de rango, varianza y desviación estándar. ¿Cómo se calculan?
# Rango es: Es el número más grande menos el número más pequeño. Entra más tamaño entonces más dispersión.
# Varianza es: Nos da una mejor idea de qué tan separados están los datos de la media.
#Aunque realmente es la dispersión de los datos respecto a la media. 
# Desviación estándar es: Es la raíz de la varianza. Lo que hace es darnos las unidades chidas. 
```

```python
# ¿Por qué es la desviación estándar una medida de dispersión más útil que el rango?
## Porque la desviación estándar nos da una mejor idea de qué tan separados están los datos de la media.
## El rango solo nos dice cuál es el número más grande y cuál es el número más pequeño, pero no nos dice cómo se distribuyen los datos respecto a la media.
```

```python
# ¿Cómo pueden dos conjuntos de datos diferentes tener la misma media pero diferentes niveles de dispersión?
## Porque la media solo nos dice el promedio de los datos, pero no nos dice cómo se distribuyen los datos respecto a la media.
##También depende de la variabilidad de los datos. La media se ve SUPER afectada por los datos 'outlayers'.

```
